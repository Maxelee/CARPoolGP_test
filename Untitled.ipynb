```python
class CARPoolGP:
    def __init__(self, Simulations, Surrogates, Kernelfunc):
        self.Simulations = Simulations
        self.Surrogates  = Surrogates
        self.Kernelfunc  = Kernelfunc
        self._build_theta_and_quantity()
        
    def _build_theta_and_quantity(self):
        
        if not Surrogates:
            self.Theta    = self.Simulations.Theta
            self.Quantity = self.Simulations.Quantity
            print('No Surrogates in this model')
        
        else:
            self.Theta    = np.concatenate((self.Simulations.Theta,    self.Surrogates.Theta))
            self.Quantity = np.concatenate((self.Simulations.Quantity, self.Surrogates.Quantity))

    def build_gp(self, GPparams):
        kernel = self.KernelFunc(GPparams)
        gp = GaussianProcess(
            kernel, self.Theta, diag=jnp.exp(GPparams["log_jitter"]), mean=GPparams["mean"]
        )
        return gp
    
    @jax.jit
    @jax.value_and_grad
    def neg_log_liklihood(self, GPparams):
        return -build_gp(GPparams).log_probability(self.Quantity)
        
    def train_gp(self, GPparams):
        kernel = self.Kernelfunc(GPparams)
        solver = jaxopt.ScipyMinimize(fun=self.neg_log_likelihood)
        soln = solver.run(GPparams, X=self.Theta, y=self.Quantity)
        return soln
```

```python
sigma_epsilon = 5 # jitter ontop of smooth variation
N=12              # Number of data samples
Domain = [-10,10] # Domain of parameters
LR = 3e-3         # Optimization learning rate 
ITERS = 3000      # max iteration
epsilon = 0.0     # Difference in smoothness from surrogates to high res
```

```python
# Toy Models
def get_Q(theta, A=-0.4, B=0.4):
    """
    This is the model which we ultimitely want to predict, represents the 
    smooth variation on some function. 
    """
    return A * theta  +  theta**3*B**4 * np.sin(theta) #+ C * x * theta + D * x

def get_R(theta, epsilon=epsilon, A=-0.4, B=0.4):
    """
    If Surrogates have some different mean function then you can make that 
    here. For simulations of same "resolution" this should return the same
    as Q
    """
    Q = get_Q(theta, A, B)
    delta = epsilon * theta  + epsilon*1/(theta**2+ epsilon)
    return Q + delta
```
